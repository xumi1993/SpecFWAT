```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
def read_seis(path, comp='Z'):
    seis = []
    for file in sorted(glob.glob(path+'/*'+comp+'.semd')):
        seis.append(np.loadtxt(file)[:,1])
    times = np.loadtxt(file)[:,0]
    return times, np.array(seis)
```

```python
def plot_seis(times, seis, comp='Z', enf=1e10, label='REF'):
    plt.figure(figsize=(4,4))
    for i in range(len(seis)):
        plt.plot(times, seis[i]*enf+i,'k')
    plt.ylabel('Trace')
    plt.xlabel('Time (s)')
    plt.title('Seismograms '+comp+' '+label)
    # plt.show()
```

```python
time_ref, seis_ref = read_seis('REF_SEIS', comp='Z')
plot_seis(time_ref, seis_ref, comp='Z', enf=1e2, label='REF')
```

```python
time_sem, seis_sem = read_seis('OUTPUT_FILES', comp='Z')
plot_seis(time_sem, seis_sem, comp='Z', enf=1e2, label='sem')
```
