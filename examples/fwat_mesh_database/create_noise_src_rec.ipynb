```python
import pandas as pd
import re
```

```python
fs = '''
FORCE  001
time shift:     0.0000
f0:             0.4
latorUTM:       62.8295
longorUTM:      -148.5509
depth:          0.0000
factor force source:             1.d15
component dir vect source E:     0.d0
component dir vect source N:     0.d0
component dir vect source Z_UP:  1.d0
'''
```

```python
sta = pd.read_csv('src_rec/STATIONS_0', header=None, sep=r'\s+')
```

```python
srcs = ['ST000', 'ST010', 'ST011', 'ST021', 'ST022', 'ST032']

with open('src_rec/sources_set0.dat', 'w') as f:
    for src in srcs:
        staname = '.'.join(sta[sta[0] == src].iloc[0, [1, 0]].values)
        stla = sta[sta[0] == src].iloc[0, 2]
        stlo = sta[sta[0] == src].iloc[0, 3]
        stel = sta[sta[0] == src].iloc[0, 4]
        f.write(f'{staname} {stla:.4f} {stlo:.4f} {stel:.2f} 0.0 1.0\n')
        stations = sta[sta[0] != src]
        stations.to_csv(f'src_rec/STATIONS_{staname}', header=None, index=None, sep=' ')  
        forceslution = re.sub(r'latorUTM:       -*\d+\.\d+', f'latorUTM:       {stla:.4f}', fs)
        forceslution = re.sub(r'longorUTM:      -*\d+\.\d+', f'longorUTM:      {stlo:.4f}', forceslution)
        with open(f'src_rec/FORCESOLUTION_{staname}', 'w') as g:
            g.write(forceslution)
```
